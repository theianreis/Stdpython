#### Operadores aritméticos
$$
\begin{array}{|c|c|c|c|}

#### Operadores de comparação
$$
\begin{array}{|c|c|c|c|}
\hline
\textbf{Operador} & \textbf{Nome} & \textbf{Exemplo em Python} & \textbf{Resultado} \\

\hline

\end{array}
