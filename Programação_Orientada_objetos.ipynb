```python
# =========================
# DEFINIÇÃO DA CLASSE
# =========================
class ContaBancaria:
    """
    Classe:
    Representa o modelo (molde) de uma conta bancária
    """

    # =========================
    # CONSTRUTOR
    # =========================
    def __init__(self, titular, saldo_inicial):
        """
        Atributos:
        - titular: nome do dono da conta
        - saldo: valor armazenado na conta
        """
        self.titular = titular        # atributo
        self.saldo = saldo_inicial    # atributo

    # =========================
    # MÉTODOS
    # =========================
    def depositar(self, valor):
        """
        Método:
        Altera o estado do objeto (saldo)
        """
        self.saldo += valor

    def sacar(self, valor):
        """
        Método:
        Remove valor do saldo, se possível
        """
        if valor <= self.saldo:
            self.saldo -= valor
        else:
            print("Saldo insuficiente")

    def mostrar_saldo(self):
        """
        Método:
        Apenas consulta o estado do objeto
        """
        print(f"Titular: {self.titular}")
        print(f"Saldo atual: R$ {self.saldo:.2f}")
```
